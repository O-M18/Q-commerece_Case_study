```python
import pandas as pd
import numpy as np
import random
from datetime import timedelta, datetime
```

```python
# constant
START_DATE='2024-01-01'
END_DATE='2024-12-31'
Date_range=pd.date_range(start=START_DATE,end=END_DATE,freq='w')
num_sku=50
categories = ['Electronics', 'Fashion', 'Home', 'Books', 'Toys']
units = ['piece', 'set', 'kg']
locations = ['Delhi', 'Mumbai', 'Bangalore', 'Hyderabad', 'Chennai']
print(Date_range)
```

```python
# Product Data
product_data = []
for i in range(num_sku):
    category = random.choice(categories)
    sku = f'SKU{i+1:03}'
    shelf_life = random.randint(30, 180)
    if category=='Home' or category=='Electronics': 
      unit = units[2]
    else:
       unit= random.choice(units[0:1])
    unit_price = round(random.uniform(100, 2000), 2)
    unit_cost = round(unit_price * random.uniform(0.5, 0.8), 2)
    product_data.append([category, sku, shelf_life, unit_price, unit, unit_cost])
product_df = pd.DataFrame(product_data, columns=['category', 'sku', 'shelf_life', 'unit_price', 'unit', 'unit_cost'])
```

```python
inventory_records = []
sales_records = []
return_records = []

for month in Date_range:
    month_no=month.month 
    for sku in product_df['sku']:
        is_problem= month_no > 6
        opening_qty=np.random.randint(50,500) 
        recieved_qty=np.random.randint(50,150) if not is_problem else np.random.randint(150,200)
        sold_qty = np.random.randint(30, 180) if not is_problem else np.random.randint(30, 120)
        closing_qty = opening_qty + recieved_qty - sold_qty
        inventory_records.append({
            'date':month,
            'sku_id':sku,
            'opening_qty':opening_qty,
            'sold_qty':sold_qty,
            'received_qty':recieved_qty,
            'closing_qty':closing_qty})
        # Sales data 
        for _ in range(np.random.randint(20,50)):
            order_id =  f'OD{_ +1:03}'
            qty = np.random.randint(1, 5)
            price = product_df[product_df['sku'] == sku]['unit_price'].values[0]
            location = random.choice(locations)
            sales_records.append({
                'order_id':order_id, 
                'sku_id':sku, 
                'date':month, 
                'quantity':qty, 
                'price':price, 
                'location':location})
        # -------- RETURN DATA GENERATION --------
            if np.random.rand() < 0.15:  # 15% chance of returns
                return_qty = np.random.randint(1, 4)
                reason = random.choice(['Damaged', 'Late Delivery', 'Wrong Item', 'Other'])
                location = random.choice(locations)
                return_records.append({
                    'retrun_id':f'R {_ +1:03}', 
                    'sku_id':sku, 
                    'date':month,
                    'retrun_id':return_qty,
                    'reason': reason, 
                    'location':location})
    
```

```python
inventory_df=pd.DataFrame(inventory_records)
sales_df=pd.DataFrame(sales_records)
retrun_df=pd.DataFrame(return_records)
```
